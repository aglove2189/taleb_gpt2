```
gpt2.download_gpt2(model_name='345M')
```

```
file_name = 'nntaleb_cleaned.txt'

sess = gpt2.start_tf_sess()

gpt2.finetune(sess,
              learning_rate=1e-5,
              dataset=file_name,
              model_name='345M',
              steps=1000,
              restore_from='fresh',
              print_every=10,
              sample_every=50,
            save_every=100
              )
```

```
gpt2.generate_to_file(sess,
                      destination_path='generated_tweets.txt',
                      length=500,
                      temperature=1,
                      nsamples=100,
                      batch_size=20
                      )
```
